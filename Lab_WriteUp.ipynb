#### Executive Summary: 
The goal of this lab was to utilize data analytics to find the best value players—those who will give us the highest return on investment and help our team improve overall. To achieve this I used clustering, which gathers unlabeled data and evaluates how similar they are to one another. This property of unlabeled data makes clustering an unsupervised Machine Learning technique. 

#### Data Preparation:
To prepare the data, I combined two datasets one with player 2025-26 salaries and another with more thorough information regarding player statistics. After merging the data, I cleaned the data by removing any duplicate rows, unnecessary columns, and missing values to ensure that the dataset is concise and the analysis is reliable. Columns removed include identifiers, awards, and player codes. I also converted salary from text to numeric format and removed dollar signs and commas since clustering algorithms only work with numeric data. This would give us a dataset of 471 players and 30 relevant statistics each.

#### Clustering Players by Performance:
After I cleaned the merged dataset, I began to select variables for clustering. I differentiated performance statistics (points, rebounds, assists, etc…) and removed text heavy variables and salary from clustering. I made this choice because I want to group players by performance patterns, not by salary or team name. Including text prior to clustering would have caused errors. We will be looking at salary later on to correspond with performance.
As mentioned before, the goal of clustering is to group players with similar performance qualities. To ensure that no one statistics variable dominates the results and skews the data, I scaled all the performance metrics so that each variable is on the same playing field. Specifically, I used StandardScalar to standardize these variables. Since KMeans measures distances between players in terms of stats, if one stat (for example, total points) has larger numbers than another (steals), it will project inaccurately as it dominates the clustering. I then applied the KMeans clustering algorithm, which grouped players based on similar qualities in performance. 
To understand the best number of clusters, I used two methods: the elbow method (how variance within clusters change with the quantity of clusters) and silhouette score (how distinct each cluster is). Add visualization here 

Based on these analyses, I chose four clusters (or K = 5) as a balance between distinction and interpretability. We could see after evaluating total within-cluster variance  and silhouette score that the highest silhouette score occurred at K = 2 (score: 0.334), which indicates strong separation between clusters. The elbow method, however, highlights the importance of having additional clusters to capture meaning data structures that would not be otherwise observed. A model with K = 5 clusters was selected as a result, having a silhouette score of 0.183 and inertia value of 5866.67—this indicates moderate separation between clusters. The fact that the clusters are not perfectly instinct makes sense as many NBA players have overlapping statistical profiles anyways.
